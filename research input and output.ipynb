```python
# Take in all of our wrestling data and read it into pandas
rd_exp = "data/Research and development expenditure (% of GDP).csv"
rd_pmp = "data/Researchers in R&D (per million people).csv"
rd_journal = "data/scientific and technical journal articles.csv"


rd_exp_df = pd.read_csv(rd_exp,skiprows=4,index_col='Country Name')
rd_pmp_df = pd.read_csv(rd_pmp,skiprows=4,index_col='Country Name')
rd_journal_df = pd.read_csv(rd_journal,skiprows=4,index_col='Country Name')
rd_exp_df.head()
```

```python
# drop columns
exp_cleaned=rd_exp_df.dropna(axis=1,how='all').drop(columns=['Country Code','Indicator Code','Indicator Name'])
pmp_cleaned=rd_pmp_df.dropna(axis=1,how='all').drop(columns=['Country Code','Indicator Code','Indicator Name'])
journal_cleaned=rd_journal_df.dropna(axis=1,how='all').drop(columns=['Country Code','Indicator Code','Indicator Name'])
exp_cleaned
```

```python
exp_cleaned2=exp_cleaned.dropna(axis=0,how='all')
pmp_cleaned2=pmp_cleaned.dropna(axis=0,how='all')
journal_cleaned2=journal_cleaned.dropna(axis=0,how='all')
exp_cleaned2
```

```python
exp_cleaned2.to_excel("exp_cleaned.xlsx")
pmp_cleaned2.to_excel("pmp_cleaned.xlsx")
journal_cleaned2.to_excel("journal_cleaned.xlsx")
```

```python
# swap rows and columns

exp_t=exp_cleaned2.transpose()
pmp_t=pmp_cleaned2.transpose()
journal_t=journal_cleaned2.transpose()
exp_t
```

```python
# save as excel
exp_t.to_excel("exp.xlsx")
pmp_t.to_excel("pmp.xlsx")
journal_t.to_excel("journal.xlsx")
```
